# Diagnostic Tests for the Best-Fit Line

### Introduction

When given a bunch of scattered points, we have to find a line which is close enough to fit these points. When finding the proper fit for a line through multiple points we need a measure which will tell us the goodness of the fit. Thus, we have to chose a 'Best Fit'. The best-fit straight line, or a linear regression line, is the line which lies in closest proximity to as many data points as possible in a given set. A linear regression line has an equation of the form $y = \alpha + \beta\ x$, where $x$ is the explanatory variable and $y$ is the dependent variable. The slope of the line is $\beta$, and $\alpha$ is the intercept (the value of $y$ when $x = 0$). Each of these statistics is determined using regression analysis in which one determines the best-fitting line (the regression line). If the coordinates of the $i^{th}$ data point are $(x_i,y_i)$, then the $y$ co-ordinate of the best-fit line at $x_i$ is $y = \alpha + \beta\ x_i$. If we have selected the appropriate values of $\alpha$ and $\beta$, then the difference $(y_i-y)$ will be small. This difference between the experimentally measured value and the theoretical prediction of our best-fit line is the pointwise vertical error $(e_i = \alpha\ x_i + \beta).$ 

To find the 'Best Fit' we have to formulate a function, which tells us the goodness of the fit, by minimizing the pointwise vertical error. It would be ideal to have this error equivalent to $0$, however, this is not typically possible given datasets with outliers. There are several different definitions for measuring and minimizing the badness-of-fit, i.e, the overall error $(E)$.

There is a standard measure to find the best fit which is called 'Least Square Fit'. Assuming that the relationship between $x$ and $y$ is a linear one, linear regression via the least squares method is the simplest approach to performing a regression analysis of a dependent and a explanatory variable. The objective is to find the best-fitting straight line through a set of points that minimizes the sum of the squared offsets from the line. This badness-of-fit is defined using the following formula:
$$E = \sum\limits_{i=1}^n e^2_i$$


Another definition used to find the badness-of-fit is: $E = \sum\limits_{i=1}^n |e_i|$, which is the sum of absolute deviation.

Our goal is to compute the best fit to $n$ points  $(x_i, y_i)$  with  $i=1,2,...n,$  via linear equation of the form
$y = \alpha + \beta\ x$, using the above given definitions of the badness of fit of a linear function to the given data. Will minimizing the squared error yield the same result as minimizing the absolute error?

We will begin by importing the modules and packages that will define and implement the functions, classes, or variables that will be used over the course of this report:

```python
from numpy import *
%matplotlib inline
import matplotlib.pyplot as plt
```

It is likely that the various definitions of overall error will disagree with most in their treatment of outliers. So some of our test cases will have data that is quite linear except for a small number of outliers.

To check how our dataset is distributed, we will produce a simple scatter plot, labelled Fig. $1$ which plots the results of a small linear data set.

```python
n = 20
x = random.rand(n)    #we have selected 20 random values for x
y = 3*x - 5           #y is calculated using the equation of line: y=mx+c, where m is 3 and c is 5.
plt.scatter(x,y);

plt.xlabel('x')
plt.ylabel('y')
plt.title('Fig. 1');
```

We will now draw a linear regression line through the data points on this scatterplot using the equation $y = \alpha + \beta\ x$.

The following function produces alpha and beta using the mean response of $y$ conditioned on $x$:

```python
def best2fit(x,y):
    xbar = sum(x)/len(x)      #mean of values in x
    ybar = sum(y)/len(y)      #mean of values in y
    n=len(x)
    
    ynum= sum([xi*yi for xi,yi in zip(x, y)]) - n * xbar * ybar #zip both arrays of x and y together when iterating through for y_i
    ydem= sum([xi**2 for xi in x]) - n * xbar**2                
    
    alpha = ynum/ydem          #slope of line
    beta = ybar - xbar*alpha   #point of intersection
    return alpha,beta

alpha,beta = best2fit(x,y)
plt.scatter(x,y)
yfit = [beta + alpha * xi for xi in x]              #line of best fit
plt.plot(x, yfit, 'r', alpha= 0.7);                 #plotting the line of best fit through the values of x

plt.xlabel('x')
plt.ylabel('y')
plt.title('Fig. 2');
```

Fig. $2$ is a regression model which perfectly minimizes the distance between the observation point and the model line. This model contains zero overall error given we used a linear data set, which did not contain any outliers. It shows positive correlation and the linear regression line passes through each data point. 

We will now test this method by adding some 'noise' to the dataset:

```python
n = 50                              #increasing size of data set
x = random.rand(n)
y = 3*x - 5 + 0.1*random.randn(n)   #adding 'noise' to y
plt.scatter(x,y)
alpha,beta = best2fit(x,y)
yfit = [beta + alpha * xi for xi in x]  #yfit is the equation for linear regression
plt.plot(x,yfit,'r',alpha= 0.7);

plt.xlabel('x')
plt.ylabel('y')
plt.title('Fig. 3');
```

The relationship displayed in Fig. $3$ is linear and shows a positive correlation with no distinct outliers. The problem with outliers is that they can have a negative effect on the regression analysis (e.g., reduce the fit of the regression equation) that is used to predict the value of the dependent (outcome) variable based on the independent (predictor) variable.

Using the least squares definition we will now fit a model to linear matrix equation:

```python
def leastsqr(x, y):
      #Computes the least-squares solution to a linear matrix equation.
    x = vstack([x, ones(len(x))]).T                    #creating a vertical array of the independent values
    return (linalg.inv(x.T.dot(x)).dot(x.T)).dot(y)    # the a single inverse matrix for the dot product of x and transpose of x
```

```python
import random

random.seed(12345)

x = [x_i*random.randrange(8,12)/10 for x_i in range(500)]          # generating a range of pseudo random numbers
y = [y_i*random.randrange(8,12)/10 for y_i in range(100,600)]

alpha, beta = leastsqr(x, y)

line_x = [round(min(x)) - 1, round(max(x)) + 1]
line_y = [alpha*x_i + beta for x_i in line_x]                      # eq for linear regression

plt.figure(figsize=(8,8));
plt.scatter(x,y);
plt.plot(line_x, line_y, color='red', lw='2');

plt.ylabel('y');
plt.xlabel('x');
plt.title('Linear regression via least squares fit');

ftext = 'y = ax + b = {:.3f} + {:.3f}x'.format(alpha, beta);    
plt.figtext(.15,.8, ftext, fontsize=11, ha='left');         # including the linear regression equation in the figure

plt.show();
plt.savefig('temp.svg');
import os
os.system('firefox temp.svg');
```

The figure displays strong positive correlation with a slope of 107.014. It appears to make efficient use of the given data. However, there might be limitations with this kind of model as the data points spread out across a longer range, with line of best fit possibly showing poor sensitivity to outliers at larger values.

We will not observe the skew of the model when the least absolute deviation method is applied:

```python
from numpy import *

def E1(alpha,beta):
    global x,y
    e = alpha*x + beta - y  # array of pointwise errors
    return abs(e).sum()     # this is E

n = 50
x = random.rand(n)
y = 3*x - 5 + 0.9*random.randn(n)

random.seed(12345)

# try a bunch of points (alpha,beta) in some box that we think contains the minimizer ...
p = 500000
alphas =  2 + 2*random.rand(p)  # in interval (2,4)
betas  = -6 + 2*random.rand(p)  # in interval (-6,-4)
# try each (alpha,beta) in turn and keep the best

Emin = inf                             # best (least) we've seen so far
for alpha,beta in zip(alphas,betas):
    E = E1(alpha,beta)
    if E < Emin:
        Emin = E
        alpha1,beta1 = alpha,beta      # best choice seen so far
plt.plot(x,y,'bo');
xx = array([0,1]);
yy = alpha1*xx + beta1 
plt.plot(xx,yy,'r');
plt.ylabel('y')
plt.xlabel('x')
plt.title('Linear regression via least absolute deviation');

ftext = 'y = ax + b = {:.3f} + {:.3f}x'.format(alpha, beta)
plt.figtext(.15,.8, ftext, fontsize=11, ha='left')  # including the linear regression equation in the figure 

plt.savefig('temp.svg')
import os
os.system('firefox temp.svg')
```

The data set used for this particular model contains large outliers, which have strongly skewed the line of best fit. It appears that variation in the data set can cause relatively large changes in the fitted plane when applying least absolute deviation.

The Least Squares definition and the Least Absolute Deviation definition may reduce the danger from outliers, but they can be unstable at data sets that are far from collinear.

## Conclusion

Both definitions appear to be strongly responsive to outliers, however, the least squares function produced a highly predicatble linear regression pattern using a very simple formula that produces quick results. Since we prefer that the regression describe the overall pattern, not the last few significant digits, in the data, unstable behavior by least absolute deviation is undesirable. When using absolute values, we will be forced to draw many different regression lines and measure the error distances from each observation to each regression line. The advantages of our least squares method are that we are sure to find only one best-fitting line. We choose to prefer the least squares method because it is much easier to work with mathematically.
